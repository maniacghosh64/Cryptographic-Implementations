```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("benchmark_results.csv")

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="size", y="clock_cycles", hue="sort", marker="o")
plt.title("Clock Cycles vs Array Size")
plt.xlabel("Array Size")
plt.ylabel("CPU Clock Cycles")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="size", y="avg_comps", hue="sort", marker="o")
plt.title("Average comparisons vs Array Size")
plt.xlabel("Array Size")
plt.ylabel("Average number of comparisons")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="size", y="constant_value", hue="sort", marker="o")
plt.title("Comparison/Complexity vs Array Size")
plt.xlabel("Array Size")
plt.ylabel("Average number of swaps")
plt.grid(True)
plt.show()
```
